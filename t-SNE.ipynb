```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.manifold import TSNE

%matplotlib inline
```

```python
import plotly.express as px
```

```python
df_main = pd.read_csv("./data/train.csv")
```

```python
df_main.head()
```

```python
pixels = df_main.columns[1:]
pixel_df = df_main[pixels]
```

```python
tsne = TSNE(verbose=5)
```

```python
tsne0 = df_main.sort_values(by="label")
```

```python
def plot_tsne(samples):
    df = tsne0[:samples]
    for i in range(1,10):
        df = pd.concat([df,tsne0.query(f"label == {i}").head(samples)],axis=0)
    labels = df.label.values
    df = df.drop(columns=['label'])
    df = tsne.fit_transform(df)
    return pd.DataFrame({
        "x":df[:,0:1].reshape(1,samples*10)[0],
        "y":df[:,1:2].reshape(1,samples*10)[0],
        "label":labels.reshape(1,samples*10)[0]
    })

    
```

### With 100 Samples

```python
df = plot_tsne(100)
px.scatter(data_frame=df,x='x',y='y',color='label')
```

### With 500 Samples

```python
df = plot_tsne(500)
px.scatter(data_frame=df,x='x',y='y',color='label')
```

### With 1000 Samples

```python
df = plot_tsne(1000)
px.scatter(data_frame=df,x='x',y='y',color='label')
```

### With 2000 samples

```python
df = plot_tsne(2000)
px.scatter(data_frame=df,x='x',y='y',color='label')
```
