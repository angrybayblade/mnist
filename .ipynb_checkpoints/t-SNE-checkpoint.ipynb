```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
import plotly.express as px
```

```python
df = pd.read_csv("./data/train.csv")
```

```python
df.head()
```

```python
pixels = df.columns[1:]
```

```python
pixel_df = df[pixels]
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(verbose=10)
```

```python
tsne0 = df.sort_values(by="label")
```

```python
samples = 500
```

```python
tsne1 = tsne0[:samples]
```

```python
for i in range(1,10):
    tsne1 = pd.concat([tsne1,tsne0.query(f"label == {i}").head(samples)],axis=0)
```

```python
labels = tsne1.label.values
```

```python
tsne1 = tsne1.drop(columns=['label'])
```

```python
tsne_df = tsne.fit_transform(tsne1)
```

```python
tsne_df = pd.DataFrame({
        "x":tsne_df[:,0:1].reshape(1,samples*10)[0],
        "y":tsne_df[:,1:2].reshape(1,samples*10)[0],
        "label":labels.reshape(1,samples*10)[0]
    })
```

```python
px.scatter(data_frame=tsne_df,x='x',y='y',color='label')
```
